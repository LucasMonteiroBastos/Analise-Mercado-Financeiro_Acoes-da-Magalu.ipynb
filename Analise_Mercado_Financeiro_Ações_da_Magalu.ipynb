```python
# bibliotecas necessarias

# libs para dataframe, modelagem e matrize
import numpy as np
import pandas as pd

# libs para analises gráficas
import matplotlib.pyplot as plt

import seaborn as sns
import plotly.graph_objects as go

# lib para ignorar avisos
import warnings

# desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# lendo os dados
database = pd.read_excel('/content/Vase_004 - Magalu - Sem Resolução.xlsx')

# verificando a base
database.head()
```

```python
# verificando o tamanho da base
database.shape
```

```python
# verificando as informações da base
database.info()
```

```python
# estatistica descritiva
database.describe()
```

```python
# series temporais

dados = database.set_index('Data')

dados.head()
```

```python
# analise das ações Magalu

# configurações do gráfico (font, tamanho, titulo, gráfico com base na data do fechamento)
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title('Analise das ações da Magalu - Fechamanto', fontsize=15, loc='center')
plt.plot(dados.index, dados['Fechamento'])

# passando o eixo x e y
plt.xlabel('Periodo da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
# final da tabela
dados.tail()
```

```python
# media movel
mediaMovel = dados['Fechamento'].rolling(5).mean()
mediaTendencia = dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,5))
plt.title('Analise das Ações da Magalu - Fechamanto', fontsize=15, loc='left')


plt.plot(dados.index, dados['Fechamento'])
plt.plot(dados.index, mediaMovel)
plt.plot(dados.index, mediaTendencia)

plt.xlabel('Periodo da Cotação')
plt.ylabel('Valor da Ação (R$)')
```

```python
# boxplot do fechamento
sns.boxplot(data=dados, x='Fechamento');
```

```python
# Boxplot Mensal

database['Mes'] = database['Data'].dt.month

plt.figure(figsize=(16,5))
sns.boxplot(data=database, x='Mes', y='Fechamento');
```

```python
# groupby mes e fechamento
database.groupby(['Mes']).describe()['Fechamento']
```

```python
# plotagem do gráfico de Candle
grafico = go.Figure(
    data=[
        go.Candlestick(
            x = dados.index,
            open = dados['Abertura'],
            high = dados['Maior'],
            low = dados['Menor'],
            close = dados['Fechamento'],

        )
    ]
)



grafico.show()
```
